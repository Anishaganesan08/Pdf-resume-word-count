```python
import os
import pytesseract
import pandas as pd
from PIL import Image
from  pdf2image import convert_from_path


#path to the directory containing the resumes
resumes_directory="/media"

#function to perform OCR on a given file and return OCR text
def perform_ocr(file_path):
    if file_name.endswith('.pdf'):
        images=convert_from_path(file_path) #convert pdf to image
        ocr_text=""
        for image in images:
            ocr_text+=pytesseract.image_to_string(image)+"\n"
    else:
        image=Image.open(file_path)
        #load image using pytesseract
        ocr_text=pytesseract.image_to_string(image)
    return ocr_text

#get the list of files in the resumes directory
resume_files=os.listdir(resumes_directory)

#create an empty dataframe to store the results
resume_data=pd.DataFrame(columns=["File Name", "OCR Text"])

#iterate over each resume file
for file_name in resume_files:
    #construct the full file path
    file_path=os.path.join(resumes_directory, file_name)

    #perform OCR on the file
    ocr_text=perform_ocr(file_path)

    #append the file name and OCR text to the dataframe
    resume_data.loc[len(resume_data)]=[file_name, ocr_text]

    #display the dataframe
    print(resume_data)
```
